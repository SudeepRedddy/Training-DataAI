```python
import os
from functools import wraps

file_path = "all_files\\sample1.txt"

def print_record_after_id(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        search_id = func(*args, **kwargs)

        with open(file_path, 'r') as lines:
            for line in lines:
                if search_id in line:
                    for _ in range(5):
                        print(line.strip())
                        try:
                            line = next(lines)
                        except StopIteration:
                            break
                    return
            print("ID not found.")
    return wrapper


@print_record_after_id
def get_id():
    return input("Enter the ID to search for: ")

get_id()
```

```python
users = {}

def register(username, password, email):
    if username in users:
        print("Username already exists!")
        return
    
    users[username] = {
        "password": password,
        "email": email
    }
    print("Registration successful!")


def login_required(func):
    def wrapper(username, password):
        if username not in users:
            print("User not found!")
            return
        
        if users[username]["password"] != password:
            print("Incorrect password!")
            return
        
        return func(username)
    return wrapper


@login_required
def dashboard(username):
    print(f"Welcome to dashboard, {username}!")



register("admin", "1234", "admin@email.com")
register("guest", "abcd", "guest@email.com")

dashboard("guest", "wrongpass")   
dashboard("guest", "abcd")        
dashboard("admin", "1234")        
```
